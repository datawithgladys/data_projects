### Revenue By Country

### Sum of Quantity and Revenue by Month
The data presents the following limitations:
1. the time stamps are repeated, i.e. multiple purchases are happening
2. the range in purchase quantity is large
3. the timescale is too granular

Hence, resampling would be done to the data based on their months, with the sum calculated for each month (for both quantity and revenue columns).


```python
#Resample by month, which contains sum of the quantities and revenue
retail_sum = retail[['quantity','revenue']].resample('M').sum()
retail_sum
```

```python
retail_sum.to_csv('./datasets/retail_sum.csv')
```

### Quantity over months
Plotting the Quantity over time will allow us to determine how many items the store sold over time, and when were the most critical period for restocks.

```python
#Plot Quantity over the months
plt.figure(figsize=(12,8))
plt.plot(retail_sum['quantity'])
plt.xlabel('Month')
plt.ylabel('Quantity')
plt.title('Quantity Over Months')
```

### Revenue over Months
Plotting the revenue over time will allow us to determine the revenue made over time and the high-performing months.

```python
#Revenue over the months
plt.figure(figsize=(12,8))
plt.plot(retail_sum['revenue'])
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Revenue Over Months')
```

## Text analysis
Here, we perform a simple text analysis to visualise and analyse the most common words to know what are the most common colours and items sold by the store.

```python
#get list of unique items sold so far and make all lowercase
unique_items = retail['description'].str.lower().unique()
unique_items
```

```python
#join the items in the list into a string 
string_items = ' '.join(unique_items)
```

```python
#Declare STOPWORDS from wordcloud library
stopwords = set(STOPWORDS)
```

### Wordcloud

```python
#Create and display wordcloud
plt.figure(figsize=(11,8))
wordcloud = WordCloud(stopwords=stopwords, max_words=50, random_state=42, background_color='white').generate(string_items)
plt.imshow(wordcloud)
plt.axis('off')
```
